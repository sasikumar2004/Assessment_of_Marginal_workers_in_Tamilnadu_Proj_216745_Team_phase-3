```python
gd.drop(['Table Code'],axis=1,inplace=True)
gd.drop(['State Code'],axis=1,inplace=True)
gd.drop(['District Code'],axis=1,inplace=True)
```

```python
gd
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
gd['Area Name'] = le.fit_transform(gd['Area Name'])
gd['Total/ Rural/ Urban']=le.fit_transform(gd['Total/ Rural/ Urban'])
gd['Age group']=le.fit_transform(gd['Age group'])
```

```python
correlation=gd.corr()
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Reds')
plt.show()
```
